```python
import pandas as pd

# List of elements
elements_list = list2

# Main dataset
data = {
    "elements": df_map.icd10.values,
    "title": df_map.title.values,
}

df = pd.DataFrame(data)

# Create an empty DataFrame to store the final results
df_final = pd.DataFrame(columns=["elements_in_the_list", "corresponded_title"], dtype=object)

# Iterate through the list of elements
for element in elements_list:
    # Check if the element exists in the main dataset
    matching_rows = df[df["elements"] == element]

    # If a match is found, append it to the final DataFrame
    if not matching_rows.empty:
        for _, row in matching_rows.iterrows():
            df_final = df_final.append({"elements_in_the_list": element, "corresponded_title": row["title"]}, ignore_index=True)

# Display the final DataFrame
print(df_final)
```

```python
import pandas as pd

entities = list5

df = df2.copy()

entity_counts = {entity: 0 for entity in entities}
entity_rows = {entity: [] for entity in entities}

for i, row in df.iterrows():
  text = row['text']

  for entity in entities:
    if entity in text:
      entity_counts[entity] += 1
      entity_rows[entity].append(i)

output = []

for entity in entities:
  data = {
    'entity': entity,
    'count': entity_counts[entity],
    'rows': ','.join(map(str, entity_rows[entity]))
  }

  output.append(data)

pd.DataFrame(output).to_csv('/Users/ilyasaden/Desktop/RunPod_workspace/data/df50_leak_text_output_v4.csv', index=False)
```
